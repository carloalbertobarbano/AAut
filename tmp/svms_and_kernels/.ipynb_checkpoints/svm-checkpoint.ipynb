```python
svc = SVC(kernel='rbf')
svc = svc.fit(X, y)
```

```python
params = svc.get_params()
V = svc.support_vectors_
print("Support vectors:", len(V))
ax = plt.gca()
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(V[:, 0], V[:, 1], s=300, linewidth=1, facecolors='none', edgecolors='black')

xlim = ax.get_xlim()
ylim = ax.get_ylim()

px = np.linspace(xlim[0], xlim[1], 30)
py = np.linspace(ylim[0], ylim[1], 30)
pY, pX = np.meshgrid(py, px)

xy = np.vstack([pX.ravel(), pY.ravel()]).T
P = svc.decision_function(xy).reshape(pX.shape)

ax.contour(pX, pY, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
ax.set_xlim(xlim)
ax.set_ylim(ylim)
    
plt.show()
```

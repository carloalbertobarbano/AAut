```python
import sklearn
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.svm import SVC

from kernel_usage_dataset import X,labels_1, labels_2, labels_3

c_1 = [1 if l == 'r' else -1 for l in labels_1]
c_2 = [1 if l == 'r' else -1 for l in labels_2]
c_3 = [1 if l == 'r' else -1 for l in labels_3]

sns.set()

sklearn.__version__
```

# Experiments with Kernels

Kernels are a powerful technique allowing one to use techniques designed to learn linear models to actually learn non-linear functions. To experiment with them, we provide you with three datasets generated using non linear functions. In these dataset each example is described by a vector $x \in \mathbb{R}^2$ while labels are from the set $\{r, b\}$.

For the sake of making the plots more legible $x$ values have been generated to fill a square grid of points in the region $[0,1] \times [0,1]$. Example descriptions have been loaded into matrix `X`. Matrix `X` is fixed for the three experiments.

Labels have been loaded in the vectors with names `labels_1`, `labels_2`, and `labels_3`.

# Plots of the three datasets

```python
plt.scatter(X[:,0].tolist(),X[:,1].tolist(), c=labels_1, marker='o')
plt.show()
plt.scatter(X[:,0].tolist(),X[:,1].tolist(), c=labels_2, marker='o')
plt.show()
plt.scatter(X[:,0].tolist(),X[:,1].tolist(), c=labels_3, marker='o')
plt.show()
```

```python
x_min, x_max = -1.5, 1.5 
y_min, y_max = -1.5, 1.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02),
                     np.arange(y_min, y_max, .02))
    
def plot_svc(svc, c):
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Pastel2, alpha=0.8)
    
    
    params = svc.get_params()
    C = svc.get_params()['C']
    print("C:", C)
    
    #svc.dual_coef_ is the product y (y=-1 or 1)
    V = svc.support_vectors_[np.abs(svc.dual_coef_[0]) < C]
    print("Support vectors:", len(V))
    
    M = svc.support_vectors_[np.abs(svc.dual_coef_[0]) == C]
    print("Margin errors:", len(M))
    
    ax = plt.gca()
    plt.scatter(X[:,0].tolist(), X[:,1].tolist(), c=c)
    plt.scatter(V[:, 0].tolist(), V[:, 1].tolist(), s=300, linewidth=2, facecolors='none', edgecolors='black')
    plt.scatter(M[:, 0].tolist(), M[:, 1].tolist(), s=300, linewidth=1, facecolors='none', edgecolors='purple')
    
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    P = svc.decision_function(xy).reshape(xx.shape)

    ax.contour(xx, yy, P, colors='k',
                   levels=[-1, 0, 1], alpha=0.5,
                   linestyles=['--', '-', '--'])
```

### Note 

To work with SVC properly your functions needs to work as follows. 

Given two matrices $X^{(1)}$ and $X^{(2)}$ containing $n$ and $m$ examples, the function returns an $n \times m$ matrix $M$ such that $M_{i,j} = K\left(X^{(1)}_i, X^{(2)}_j\right)$ (where $X^{(k)}_i$ is the i-th example in matrix $X^{(k)}$).



### Note 2

See the slides titled "Important Kernels" for a list of possible kernels to use.

### Note 3

Sometimes support vector machines are a little too aggressive in regularising the resulting function. In some of these exercises you might need to experiment with different values of `C` to avoid this problem.

```python
def kernel_1(sigma): #gaussian kernel (rbf)
    # exp(-||Xi - Xj||/2sigma^2)
    def K(X, Y):
        norm = np.dot(X, -Y.T)
        return np.exp(-norm/(2*sigma**2))
    return K

def kernel_2(exp): #polinomial kernel of degree d
    def K(X, Y):
        M = np.dot(X, Y.T)
        M = np.power(M+1, exp)
        return M
    return K

def kernel_3(exp, sigma):
    return lambda X, Y: kernel_2(exp)(X, Y) * kernel_1(sigma)(X, Y)
```

```python
svc = SVC(kernel=kernel_1(sigma=0.5))
svc = svc.fit(X, c_1)
svc.support_vectors_ = X[svc.support_, :]

plt.figure(figsize=(18, 9))
plot_svc(svc, labels_1)
plt.show()
```

## Polynomial kernel of deegree d (dataset 2)

```python
svc = SVC(kernel=kernel_2(exp=3), C=1000)
svc = svc.fit(X, c_2)
svc.support_vectors_ = X[svc.support_, :]

plt.figure(figsize=(18, 9))
plot_svc(svc, labels_2)
plt.show()
```

## Combination of kernels

```python
svc = SVC(kernel=kernel_3(exp=3, sigma=0.1), C=10.)
svc = svc.fit(X, c_3)
svc.support_vectors_ = X[svc.support_, :]


plt.figure(figsize=(18, 9))
plot_svc(svc, labels_3)
plt.show()
```

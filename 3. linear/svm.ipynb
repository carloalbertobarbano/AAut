### Part 1:

- import the support vector machine classifier from scikit-learn (the SVC class) and train a classifier for the examples above using a linear kernel;
- read the documentation to find out how to obtain the support vectors and the associated (dual) weights; use this information to analyze the learnt model: 
    - how many support vectors have been learnt? 
    - are them in the position you would have expected [[1](#note1)]? 
    - is there any margin error?
    - is there any classification error (check it using the classifier predictions)?

    
<a name="note1">[1]</a> If you make two plots one after the other (in the same cell), the plots will be merged into a single one. You may want to use this feature to plot the support vectors on top of the scatter plot for the dataset.

```python
svc = SVC(kernel='linear')
svc = svc.fit(X, Y)
```

```python
x_min, x_max = -1.5, 1.5 
y_min, y_max = -1.5, 1.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, .002),
                     np.arange(y_min, y_max, .002))
    
def plot_svc(svc):
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Pastel2, alpha=0.8)
    
    
    params = svc.get_params()
    C = svc.get_params()['C']
    print("C:", C)
    
    #svc.dual_coef_ is the product yα (y=-1 or 1)
    V = svc.support_vectors_[np.abs(svc.dual_coef_[0]) < C]
    print("Support vectors:", len(V))
    
    M = svc.support_vectors_[np.abs(svc.dual_coef_[0]) == C]
    print("Margin errors:", len(M))
    
    ax = plt.gca() 
    plt.scatter(X[:,0], X[:,1], c=c)
    plt.scatter(V[:, 0], V[:, 1], s=300, linewidth=2, facecolors='none', edgecolors='black')
    plt.scatter(M[:, 0], M[:, 1], s=300, linewidth=1, facecolors='none', edgecolors='purple')
    
    xy = np.vstack([xx.ravel(), yy.ravel()]).T
    P = svc.decision_function(xy).reshape(xx.shape)

    ax.contour(xx, yy, P, colors='k',
                   levels=[-1, 0, 1], alpha=0.5,
                   linestyles=['--', '-', '--'])

plt.figure(figsize=(18, 9))
plot_svc(svc)
plt.show()
```

**R**:
- There are 17 support vectors (3 real support vectors (α=C) and 14 margin errors)
- There is no classification error
- Support vectors are in the expected position given C=1

### Part 2:

**Q**:
- learn a new SVC model using custom C values:
    - how the answers to the questions above change when you use a very high C value (e.g., 1000)?
    - how the answers to the questions above change when you use a very low C value (e.g., 0.3)?

```python
svc_low = SVC(kernel='linear', C=0.3)
svc_low = svc_low.fit(X, Y)

svc_high = SVC(kernel='linear', C=1e3)
svc_high = svc_high.fit(X, Y)

plt.figure(figsize=(18, 9))

plt.subplot(1, 2, 1)
plt.title('C=0.3')
plot_svc(svc_low)

plt.subplot(1, 2, 2)
plt.title('C=1000')
plot_svc(svc_high)
plt.show()
```

**R**:
    - again no classification error
    - Yes, positions and number of support vectors/margin errors are expected given the values of C
      0.3 and 1000

### Part 3

**Q**:
- learn a new SVC model using a rbf kernel:
    - is the new kernel able to capture the linear model?
    - are you surprised by the above answer? why you are (not) surprised?

```python
svc_rfb = SVC(kernel='rbf', C=10.0, gamma='auto')
svc_rfb = svc_rfb.fit(X, Y)

svc_rfb_low = SVC(kernel='rbf', C=0.1, gamma='auto')
svc_rfb_low = svc_rfb_low.fit(X, Y)

plt.figure(figsize=(18, 9))

plt.subplot(1, 2, 1)
plt.title('C=10.0')
plot_svc(svc_rfb)

plt.subplot(1, 2, 2)
plt.title('C=0.1')
plot_svc(svc_rfb_low)
plt.show()
```

**R**:
    - Yes, the gaussian kernel is able to solve the linear task
    - Not surprised (but changes to C produce interesting results, i.e C=0.1 produces classification errors, 
      but again not surprising since the regularization on the allowed margins is lower (in the dual we 
      constrain α to a smaller value and thus reduce the weight of all the samples -> optimal value of C is a
      tradeoff between achieving a good classification and avoid overfitting due to possible outliers)
